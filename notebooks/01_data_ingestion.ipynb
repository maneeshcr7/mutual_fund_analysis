# 01 — Data Ingestion

Load and explore all 10 CSV datasets. Validate schema, check for anomalies.

```python
import pandas as pd
import numpy as np
from pathlib import Path
import os

os.chdir('/Users/maneeshreddy/Desktop/mutual_fund_analysis')
DATA_DIR = Path('data_1/Bluestock_MF_Datasets')

CSV_FILES = [
    '01_fund_master.csv', '02_nav_history.csv', '03_aum_by_fund_house.csv',
    '04_monthly_sip_inflows.csv', '05_category_inflows.csv',
    '06_industry_folio_count.csv', '07_scheme_performance.csv',
    '08_investor_transactions.csv', '09_portfolio_holdings.csv',
    '10_benchmark_indices.csv'
]

print(f'Data directory: {DATA_DIR}')
print(f'Files to load: {len(CSV_FILES)}')
```

```python
datasets = {}
for f in CSV_FILES:
    fp = DATA_DIR / f
    if fp.exists():
        df = pd.read_csv(fp)
        datasets[f] = df
        print(f'\n{"="*60}')
        print(f'{f}: {df.shape[0]} rows x {df.shape[1]} cols')
        print(f'Columns: {list(df.columns)}')
        print(f'Missing: {df.isnull().sum().sum()}')
        print(f'Duplicates: {df.duplicated().sum()}')
    else:
        print(f'\nMISSING: {f}')

print(f'\n{"="*60}')
print(f'Loaded {len(datasets)}/10 datasets')
```

```python
# Quick profile of each dataset
for name, df in datasets.items():
    print(f'\n{"─"*50}')
    print(f'📊 {name}')
    print(f'{"─"*50}')
    display(df.head(3))
    print(f'\nDtypes:\n{df.dtypes.to_string()}')
```
